```python
import numpy as np

# Function to generate a random user profile with masked float points
def generate_user_profiles(num_profiles, num_features):
    # Generate random user profiles
    user_profiles = np.random.rand(num_profiles, num_features)
    
    # Masking the user profiles by normalizing them to have values between 0 and 1
    # This is a simple form of masking, and more complex methods could be employed if needed
    user_profiles = user_profiles / np.linalg.norm(user_profiles, axis=1, keepdims=True)
    
    return user_profiles

# Generate 10 user profiles with 300 features each, which is a common dimensionality for vectors
num_profiles = 10
num_features = 300
user_profiles = generate_user_profiles(num_profiles, num_features)

masked_user_profiles = {f"user_{i}": user_profiles[i].tolist() for i in range(num_profiles)}
```
